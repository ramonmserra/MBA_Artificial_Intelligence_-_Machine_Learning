### Enunciado

A partir dos dados housing.csv, realize toda análise exploratória necessária e prepare os dados para algoritmos de machine learning usando a Scikit-Learn. Treine um modelo de regressão linear que prediga a variável meta (median_house_value) e cujo valor de RMSE seja menor ou igual a 69000. 

```python
import pandas as pd
import numpy as np

# leitura dos dados
housing = pd.read_csv('bases/housing.csv')
housing.head()
```

```python
housing.describe()

#housing=housing.mask((housing - housing.mean()).abs() > 2 * housing.std()).dropna()
```

```python
housing.groupby('ocean_proximity').size()

ONEH_OCEAN= housing[housing['ocean_proximity']=='<1H OCEAN']
ONEH_OCEAN = ONEH_OCEAN['median_house_value'].median()
INLAND = housing[housing['ocean_proximity']=='INLAND']
INLAND = INLAND['median_house_value'].median()
ISLAND = housing[housing['ocean_proximity']=='ISLAND']
ISLAND = ISLAND['median_house_value'].median()
NEAR_BAY = housing[housing['ocean_proximity']=='NEAR BAY']
NEAR_BAY = NEAR_BAY['median_house_value'].median()
NEAR_OCEAN = housing[housing['ocean_proximity']=='NEAR OCEAN']
NEAR_OCEAN = NEAR_OCEAN['median_house_value'].median()
ocean_mapping = {'<1H OCEAN': ONEH_OCEAN,'INLAND':INLAND,'ISLAND': INLAND, 'NEAR BAY' : NEAR_BAY, 'NEAR OCEAN': NEAR_OCEAN }

housing['ocean_proximity'] = housing['ocean_proximity'].map(ocean_mapping)

housing.head()


```

```python
housing['total_bedrooms'] = housing['total_bedrooms'].fillna(housing['total_bedrooms'].median())

housing.describe()
```

```python
housing_norm = housing.copy()

housing_norm = (housing_norm - housing_norm.mean())/housing_norm.std()

housing_norm.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error

X = housing.drop('median_house_value',axis=1)

y = housing['median_house_value']

split_test_size = 0.20
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split_test_size)
```

```python
lr_model = linear_model.LinearRegression()
lr_model.fit(X_train, y_train.ravel()) 
```

```python
y_pred = lr_model.predict(X_test)
```

```python
import math
math.sqrt(mean_squared_error(y_test,y_pred))
```
